# Plot a plot inside a plot with inset_axes.

Basically: How to zoom part of a graph? But you could also do more than just that though.

While experimenting I've found it easier to use the recommended path by matplotlib, but to set all params by hand instead of letting mpl decide.

Anyhow I've got some Walkers from Stochastic simulations done in C++ for uni class. For each Waler I was supposed to calculate average values of momentum to finally get some average energy values. Completely really irrelevant actually....but it pays out to check the DataFiles folder to see how the file is structured and read in: 

```python
def readCSV(filename):
    
    f = open(filename)
    readlines = csv.reader(open(filename))
    
    block, RBlock, RAll = [], [], []
    title = ""
    for row in f.readlines():
        if row.startswith("#"):
            title+=row
        else:
            line = row.split(",")
            try:
                block.append(float(line[0]))
                RBlock.append(float(line[1]))
                RAll.append(float(line[2]))
            except:
                pass

    title = title.replace(",", " ")
    title = title.replace("\n", "")
    title = title.replace("#", "")
    return title, block, RBlock, RAll
```

Anyhow now we have our plot title, x axis set by "blocks" and then y values set by RBlock and Rall.
Plotting time...

```python
def plotCSV(filename, returned=False):
            
    title, block, RBlock, RAll = readCSV(filename)
    fig, ax = plt.subplots(figsize=[16,7])

    #MAIN PLOT SETTINGS AND PLOTS
    #----------------------------------------------------------------------
    ax.plot(block, RBlock, label="<P> per blocks", linewidth=2)
    ax.plot(block, RAll, label="<P> per total counts", linewidth=2)
    plt.title(title)
    plt.xlabel("N")
    plt.ylabel("<P>")
    

    #INSERTED AXIS SETTINGS AND PLOTS
    #up to this point things were relatively normal. Now, we insert new
    #axes with inset_axes with params: (parent axis, dimensions, transformation)
    #----------------------------------------------------------------------
    axins = inset_axes(ax, width=5, height=3, bbox_to_anchor=(0.7, 0.7),
                 bbox_transform=ax.figure.transFigure)
    
    #these just plot the data from the file onto the graph. I get min max values and 
    #set the ranges and line parameters as I need them. Nothing really important
    axins.plot(block, RBlock, label="<P> per blocks", linewidth=1.5)
    axins.plot(block, RAll, label="<P> per per total count", linewidth=1.5) 
    x1, x2, y1, y2 = block[-1]-1, block[-1], RAll[-1]-0.1, RBlock[-1]+0.1
    axins.set_xlim(x1, x2)
    axins.set_ylim(y1, y2)
    
    #This line prevents mpl from canging the axis offset. Ever saw 
    #that nasty writing like: +3.886e2 on the top of your graph?
    #and then all your y-values change to 0.0001 to 0.00005? Well this line kills that
    #and forces mpl to write FULL numbers on your y axis! In the example above that's: 388.7
    plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
    
    #this line draws the gray lines from our inserted graph to the area that's being zoomed into.
    mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")
    
    #as you might've noticed above, when you issue a plt.command AFTER you've already 
    #used the inset_axes command, ALL PLT COMMANDS NOW REFER TO IT (axins in our case)
    #this adds labels for the inserted graph. If you want to add something to main graph,
    #with plt.command just add the command before inset_axes
    #Notice how I moved the legend lower so the lines are still visible, again, by using bbox_to_anchor.
    plt.xlabel("N")
    plt.ylabel("<P>")
    plt.legend(bbox_to_anchor=(0.9, 0.85),
               bbox_transform=ax.figure.transFigure)
    

    #PLOT IT
    #----------------------------------------------------------------------
    plt.show()
    
    if returned:
        return block, RBlock, RAll
    

plotCSV(r"/DataFiles/inline_plot.txt")
```

Ok so as comments said, parameters inset_axes takes are: (parent axis, dimensions, location, transformation)

<ul>
<li>__Parent axis__: mandatory. These are the axis mpl looks at while drawing your graph. Each time you draw it differently, mpl recalculates the relative sizes and positions of all elements it draws. Because we're trying to plot something that's inescapably related to another object mpl needs that reference to know "in relation to what" am I evaluating this.
<li>__Dimension__: pretty straight forward, give it either a relative size as in height/width = XX% (relative to parent axes that is). Or provide it with "exact" figure sizes like I did above.
<li>__location__: Ah now this is the 1st tricky part. You can tell mpl the location by using a "loc" keyword. But I found it much, __much__ easier to guess the location I like with bbox_to_anchor command. Node that you can have both commands at the same time. I don't think it's meant to be used that way because then you often run into graph scaling problems. Zooming in breaks it etc... The automatic thing also behaves weirdly so it's hard to figure out what exactly is going on. I.e. check out how the positions of the axins go:
</ul>

```python
fig, ax = plt.subplots(figsize=[20,8])
ax.set_axis_bgcolor("#FFAEAE")

def drawAxins(loc, text):
    axin = inset_axes(ax, width=2, height=1.5, loc=loc)
    axin.text(0.43,0.4, text, fontsize=24, family="cursive", weight="bold")
    plt.gca().set_axis_bgcolor("lightblue")


#INSERTED AXIS SETTINGS AND PLOTS
#up to this point things were relatively normal. Now, we insert new
#axes with inset_axes with params: (parent axis, dimensions, transformation)
#----------------------------------------------------------------------
drawAxins(1, "1")
drawAxins(2, "2")
drawAxins(3, "3")
drawAxins(4, "4")
drawAxins(5, "5")
drawAxins(6, "6")
#drawAxins(7, "7") 7 OVERLAPS 5!!!!!
drawAxins(8, "8")
drawAxins(9, "9")
drawAxins(10, "10")

plt.show()
```

Do you get the confusion now? Also these are just "general" guidelines, they're hardly precise, as you can see, they like to overlap with everything and make a mess. 
Like I said much much easier to just use the relative coordinates with bbox_to_anchor, it's the upper left corner I beleive that's being positioned, and place your graph on a trial-and-error basis. But when you use the bbox, you have to give it a manner of transformation.

If you don't, then scaling your graph, will produce results as weird as with the loc keyword. That's why you MUST NOT forget to add the param: bbox_transform=ax.figure.transFigure

##### mark_inset: 

The last thing on the list of to-do's that makes our graph cool.
This command draws the gray lines that lead to the zoomed area. You send it the parent axes object, the zoomed in ("child") axes object, which is in our case axins, and you send it 2 integers. That's the mandatory part. 

Integers go from 1 to 4 and represent CORNERS which the lines connect. Counter-clockwise like the loc keyword above (see the 1,2,3,4 axins in graph above). The "extra stuff": fc="none", ec="0.5" just sets facecolor of the line to none and edgecolor to 0.5 to get a nice non-intrusive gray line color.



If you don't write it down you're not going to remember it so don't be lazy.
