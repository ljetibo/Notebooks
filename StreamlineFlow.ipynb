```python
import csv
import os

path = r"C:\Users\User\Notebooks\DataFiles\flow"

def ReadCSV(path_to_dat):
    x, y = list(), list()
    dataf = open(path_to_dat, "r")
    for line in dataf.readlines():
        x.append(float(line.split(" ")[0]))
        y.append(float(line.split(" ")[1]))
    return x, y

def SortData(path):
    xax, yax, x, y = list(), list(), list(), list()
    for root, dirs, files in os.walk(path):
        for filename in files:
            xax, yax = ReadCSV(os.path.join(root, filename))
            x.append(xax)
            y.append(yax)
    return x, y

x, y = SortData(path)
len(x)
```

```python
class charge:
    def __init__(self, q, pos):
        self.q=q
        self.pos=pos
 
        

def E_point_charge(q, a, x, y):
    """
              1       qi*ri
    Ei = ---------- * --------
         4*pi*eps0     ri^2

    Separate ri into xi^2+yi^2 and run xi and yi across the desired ploting area
    This calculates values in those points in space, return arrays for Exi and Eyi
    with E field strenght values at given points xi, yi
    I.e. For Ex you get in carhesian coords:
    
                  q*rx            |where rx is distance of a point on the graph
    Ei =  ---------------------   |and location of the point charge, basically:
            (rx^2 + ry^2)^(3/2)   |x_plot - x_ofPointCharge
                                  ndarray    fixed point

    """
    return q*(x-a[0])/((x-a[0])**2+(y-a[1])**2)**(1.5), \
        q*(y-a[1])/((x-a[0])**2+(y-a[1])**2)**(1.5)       
 
      
def E_total(x, y, charges):
    """
                  1    SUM   qi*ri
    E = ------------ SUM    -------
         4*pi*eps0     SUM   ri^2
                        i
    Run through all sent charges, and add E field values in all the points
    on the graph into an ndarray for final total value of E field.
    """    
    Ex, Ey=0, 0
    for C in charges:
        E=E_point_charge(C.q, C.pos, x, y)
        Ex=Ex+E[0]
        Ey=Ey+E[1]
    return [ Ex, Ey ]

def E_dir(t, y, charges):
    Ex, Ey=E_total(y[0], y[1], charges)
    n=sqrt(Ex**2+Ey*Ey)
    return [Ex/n, Ey/n]

# charges and positions
#charges=[ charge(1, [-0.15, 0]), charge(-1, [0.15, 0]) ]
#QUADRIPOLE
#charges=[ charge(-1, [0.2, 0.2]), charge(-1, [-0.22, -0.2]), 
#         charge(1, [-0.2, 0.2]), charge(1, [0.22, -0.2]) ]
#OCTUPOLE
charges=[charge(-1, [0.5, 0.5]), charge(-1, [-0.5, -0.5]), 
         charge(-1, [-0.5, 0.5]), charge(-1, [0.5, -0.5]),
         charge(1, [0, 0.5]), charge(1, [0, -0.5]), 
         charge(1, [-0.5, 0]), charge(1, [0.5, 0]) ]


# plot field lines
x0, x1=-1.5, 1.5
y0, y1=-1.5, 1.5
R=0.001
# loop over all charges
for C in charges:
    # plot field lines starting in current charge
    dt=0.8*R
    if C.q<0:
        dt=-dt
    # loop over field lines starting in different directions 
    # around current charge
    for alpha in linspace(0, 2*pi*15/16, 16):
        r=ode(E_dir)
        r.set_integrator('vode')
        r.set_f_params(charges)
        x=[ C.pos[0] + cos(alpha)*R ]
        y=[ C.pos[1] + sin(alpha)*R ]
        r.set_initial_value([x[0], y[0]], 0)
        while r.successful():
            r.integrate(r.t+dt)
            x.append(r.y[0])
            y.append(r.y[1])
            hit_charge=False
            # check if field line left drwaing area or ends in some charge
            for C2 in charges:
                if sqrt((r.y[0]-C2.pos[0])**2+(r.y[1]-C2.pos[1])**2)<R:
                    hit_charge=True
            if hit_charge or (not (x0<r.y[0] and r.y[0]<x1)) or \
                    (not (y0<r.y[1] and r.y[1]<y1)):
                break
        plt.plot(x, y, '-k')
 
# plot point charges
for C in charges:
    if C.q>0:
        plot(C.pos[0], C.pos[1], 'bo', ms=8*sqrt(C.q))
    if C.q<0:
        plot(C.pos[0], C.pos[1], 'ro', ms=8*sqrt(-C.q))
          

gca().set_xlim(x0, x1)
gca().set_ylim(y0, y1)
plt.show()
```
