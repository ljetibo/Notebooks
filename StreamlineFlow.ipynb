```python
import csv
import os

path = r"C:\Users\Dino\Desktop\Test\Notebooks\DataFiles\flow"

def ReadCSV(path_to_dat):
    x, y = list(), list()
    dataf = open(path_to_dat, "r")
    for line in dataf.readlines():
        x.append(float(line.split(" ")[0]))
        y.append(float(line.split(" ")[1]))
    return x, y

def SortData(path):
    xax, yax, x, y = list(), list(), list(), list()
    for root, dirs, files in os.walk(path):
        for filename in files:
            xax, yax = ReadCSV(os.path.join(root, filename))
            x.append(xax)
            y.append(yax)
    return x, y

x, y = SortData(path)
```

```python
class charge:
    """
    Class representing a single charged particle. 
    charge class is consisted of a charge value q
    and a tuple representing coordinates of the charged particle.
    """
    def __init__(self, q, pos):
        """
        Send (q, pos) where q is a float value representing charge
        and pos is a tuple of coordinates (x,y)
        """
        self.q=q
        self.pos=pos
 
def E_point_charge(q, a, x, y):
    """
              1       qi*ri
    Ei = ---------- * --------
         4*pi*eps0     ri^2

    Separate ri into xi^2+yi^2 and run xi and yi across the desired ploting area
    This calculates values in those points in space, return arrays for Exi and Eyi
    with E field strenght values at given points xi, yi
    I.e. For Ex you get in carhesian coords:
    
                  q*rx            |where rx is distance of a point on the graph
    Ei =  ---------------------   |and location of the point charge, basically:
            (rx^2 + ry^2)^(3/2)   |x_on_plot - x_ofPointCharge
                                  
    If you read the code carefully, you'll see that you're subtracting: ndarray-number
    when you're calculating ri. This can happen because numpy will automatically subtract
    the number from all values inside a numpy array. This is a quick way of getting all
    ri's on our graph.
    """
    return q*(x-a[0])/((x-a[0])**2+(y-a[1])**2)**(1.5), \
        q*(y-a[1])/((x-a[0])**2+(y-a[1])**2)**(1.5)       
 
      
def E_total(x, y, charges):
    """
             1       ---   qi*ri
    E = ------------ \    -------
         4*pi*eps0   /     ri^2
                     ---     
                      i
    Run through all sent charges, and add E field values in all the points
    on the graph into an ndarray for final total value of E field.
    """    
    Ex, Ey=0, 0
    for C in charges:
        E=E_point_charge(C.q, C.pos, x, y)
        Ex=Ex+E[0]
        Ey=Ey+E[1]
    return [ Ex, Ey ]


def plotCharges(charges, xspan=(-2,2), yspan=(-2,2), npoints=2000):
    """
    Function that finally draws the electric field. Send charges
    as a list of charge objects of all charges you want to plot"""
    #get a location grid in place.
    x= np.linspace(xspan[0], xspan[1], npoints)
    y= np.linspace(yspan[0], yspan[1], npoints)
    
    #get actual points in which we'll calc our E field.
    x, y = meshgrid(x, y) #create a meshgrid from the 2k equaly spaced points 
                          #basically, match appropriate x's and y's
        
    Ex, Ey=E_total(x, y, charges) #calc the actual field values 

    streamplot(x, y, Ex, Ey, color='k') #and draw the lines

    # plot point charges (circles/dots)
    for C in charges:
        if C.q>0:
            plot(C.pos[0], C.pos[1], 'bo', ms=8*sqrt(C.q))
        if C.q<0:
            plot(C.pos[0], C.pos[1], 'ro', ms=8*sqrt(-C.q))

    gca().set_xlim(xspan[0], xspan[1])
    gca().set_ylim(yspan[0], yspan[1])
    plt.show()

    
# charges and positions
#DIPOLE
charges=[ charge(1, [0.005, 0]), charge(-1, [-0.005, 0])]
plotCharges(charges)
```

The thing that's really bothering me, especially with the Octupoles is that the field itself is not very visible. Lines are broken up and don't follow the standard physical convention that they have to start and end in a charged particle.

```python
def E_deriv(t, y, charges):
    Ex, Ey=E_total(y[0], y[1], charges)
    n=sqrt(Ex**2+Ey**2)
    return [Ex/n, Ey/n]


def SolveNPlot(func, charges, xspan=(-2,2), yspan=(-2,2), pointstep=0.001):
    for C in charges:
        # plot field lines starting in current charge
        dt=0.8*pointstep
        if C.q<0:
            dt=-dt

        # loop around current charge C and integrate  
        # loop over field lines starting in different directions 
        # to get the next step
        for alpha in linspace(0, 2*pi*15/16, 16):
            #set up the ODE func that needs to be solved
            #this is the func that'll get integrated
            r=ode(func)
            r.set_integrator('vode')
            r.set_f_params(charges)

            x=[ C.pos[0] + cos(alpha)*pointstep ]
            y=[ C.pos[1] + sin(alpha)*pointstep ]
            r.set_initial_value([x[0], y[0]], 0)

            while r.successful():
                #integrate to get every next step 
                r.integrate(r.t+dt)
                x.append(r.y[0])
                y.append(r.y[1])
                hit_charge=False
                # check if field line left drawing area or ends in some charge
                # if the next step has landed on a charge, or has exited the 
                # drawing area, stop integrating
                for C2 in charges:
                    if sqrt((r.y[0]-C2.pos[0])**2+(r.y[1]-C2.pos[1])**2)<pointstep:
                        hit_charge=True
                if hit_charge or (not (xspan[0]<r.y[0] and r.y[0]<xspan[1])) or \
                        (not (yspan[0]<r.y[1] and r.y[1]<yspan[1])):
                    break
            #plot whatever is located in the x and y 
            plt.plot(x, y, '-k')
 
        # plot point charges
        for C in charges:
            if C.q>0:
                plot(C.pos[0], C.pos[1], 'bo', ms=8*sqrt(C.q))
            if C.q<0:
                plot(C.pos[0], C.pos[1], 'ro', ms=8*sqrt(-C.q))

        gca().set_xlim(xspan[0], xspan[1])
        gca().set_ylim(yspan[0], yspan[1])
    plt.show()
```

```python
# charges and positions
#DIPOLE
charges=[ charge(1, [-0.15, 0]), charge(-1, [0.15, 0]) ]
SolveNPlot(E_deriv, charges)
```

```python
#QUADRIPOLE
charges=[ charge(-1, [0.2, 0.2]), charge(-1, [-0.22, -0.2]), 
         charge(1, [-0.2, 0.2]), charge(1, [0.22, -0.2]) ]
SolveNPlot(E_deriv, charges)
```
