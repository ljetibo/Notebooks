<h1>Dave's at it again</h1>

Solution for Dave Beazley's Dave's bus problem from: http://www.dabeaz.com/pydata/.
Find a bus Dave could've potentially left his suitcase on. 
General approach:
<ul>
<li>Get a list of busses on route 22 in Chicago.</li>
<li>Save that list for in a file, just in case.</li>
<li>Get all busses north from Dave's office (41.980262, -87.668452) </li>
<li>Get those that are traveling North</li>
<li>Get just those that are nearer than 1km</li>
<li>Show them on Google Maps using Google Maps API</li>
</ul>

Implemented also is an active tracking feature which updates every ~15mins by default and displays new busses.

```python
from urllib.request import urlopen #grabs the busses!
import math
from xml.etree.ElementTree import parse #parse xml into a tree of xml elements!
import webbrowser #for google API 
```

```python
def deg2rad(deg):
    """Convert degrees to radians. Trigonometric
    functions operate only on radians."""
    
    return deg*math.pi/180

def eq2m(lat, lon):
    """Converts sent latitude and longitude into meters. 
    Make sure you send differences of lat and lon, otherwise you'll get distance
    from Grenwich to (lat, lon)"""
    
    R = 6378.1
    return (R*deg2rad(lat), R*lon*math.cos(deg2rad(lat)))

class Location:
    """Set desired location. Get back
    latitude and longitude in degrees and radians"""
    
    def __init__(self, lat, lon):
        self.lat = lat
        self.lon = lon
        self.rlat = deg2rad(lat)
        self.rlon = deg2rad(lon)

def GetDataLocally(url, filename):
    """Open the url of api (***any api***) and retrieve data requested
    with the sent link. Save recieved data into a file.
    Returns filename (includes filepath)"""
    
    u = urlopen(url)
    data = u.read()
    f = open(filename, "wb")
    f.write(data)
    f.close()
    return filename


def findbus(busses, location):
    """Returns a list of xml elements that are north of set
    location, are going north, and are closer than 1km.
    """
    #Parse the grabbed file into a tree then run down it 
    #separating the busses you need
    tree = parse(busses)
    buslist = list()
    
    for bus in tree.findall("bus"):
        d=bus.findtext("d")
        lat = float(bus.findtext("lat"))
        lon = float(bus.findtext("lon"))
        
        diff = eq2m(location.lat-lat, location.lon-lon)
        
        if (lat > location.lat):
            if (d.startswith("North")):
                if math.sqrt(diff[0]**2 + diff[1]**2) < 1000:
                    buslist.append(bus)
                    
    return buslist


def ShowBussesOnMap(location, url, filename="rt22.xml"):
    """Uses Google Maps API to show location of potential busses Dave 
    might have left his suitcase on. Optional send variable is filename,
    that is path and name of file where downloaded data will be saved.
    location has to be an instance of Location class
    url is the link to Chicago buss routes API."""
    
    all_busses = GetDataLocally(url, filename)
    busses = findbus(all_busses, location)
    baseurl = "http://maps.googleapis.com/maps/api/staticmap?&zoom=11&size=400x400&markers=color:red|"
    
    for bus in busses:
        baseurl += bus.findtext("lat")+","+bus.findtext("lon")+"|"
    webbrowser.open_new_tab(baseurl)

    
#run it:
Dave = Location(41.980262, -87.668452)
ChicagoBusRoutesAPI = "http://ctabustracker.com/bustime/map/getBusesForRoute.jsp?route=22"
ShowBussesOnMap(Dave, ChicagoBusRoutesAPI)

def ActiveTrack(location, url, filename="rt22.xml"):
    """Uses Google Maps API to show location of potential busses Dave 
    might have left his suitcase on. Actively tracks their locations,
    and updates them on the map."""
    import time
    all_busses = GetDataLocally(url, filename)
    busses = findbus(all_busses, location)
    baseurl = "http://maps.googleapis.com/maps/api/staticmap?&zoom=11&size=400x400&markers=color:red|"
    
    while True:
        baseurl1 = baseurl
        for bus in busses:
            baseurl1 += bus.findtext("lat")+","+bus.findtext("lon")+"|"
        webbrowser.open(baseurl1)
        time.sleep(1000)

ActiveTrack(Dave, ChicagoBusRoutesAPI)
```
